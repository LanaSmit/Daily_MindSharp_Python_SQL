```python
#General 

#Buidling an ETL Data Pipeline

#Step 1: Extract from CSV file into a Pandas Dataframe
#Step 2: Transform the data( clean the data) - Deal with missing and duplicate data 
#Step 3:Create a database
#Step: Load the clean data into the database

import pandas as pd
import psycopg2 # For comectimg python to postresql database
from sqlalchemy import create_engine # To efficiently manage and reuse the database connections

data = pd.read_csv("Softwork.csv")
data.head()
```

```python
data.duplicated().sum()
```

```python
#Remove duplicate records
data.drop_duplicates(keep='first', inplace=True)
data.duplicated().sum()
```

```python
data.duplicated().sum()
```

```python
# Explore missing data
data.isnull().sum()
```

```python
data[data['education'].isnull()].head()
```

```python
# Deal with missing data
data['education'].fillna('unknown', inplace=True)
data['previous_year_rating'].fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

```python
#Create a database - PGAdmin and create a database table
#Database Credentials 
username = 'postgres'
password = 'lana'
host = 'localhost'
port = 5432
db_name = 'postgres'
```

```python
# Extablish a connection
engine = create_engine(f'postgresql://{username}:{password}@{host}:{port}/{db_name}')
```

```python
# Load the database table  - employee_table
data.to_sql('employee_table', engine, if_exists='replace',index=False)

#Close
engine.dispose()
```
