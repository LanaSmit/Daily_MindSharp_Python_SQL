```python
# 3. LeetCode : Longest Substring without repeating characters
listNoDupLetter=[]
input="pwwkew"
substring=''
for i in input:
    if i not in listNoDupLetter:
        listNoDupLetter.append(i)
        substring= substring+i
        

SubstringLength= len(listNoDupLetter)
print(substring)
print(SubstringLength)
```
