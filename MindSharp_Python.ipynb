```python
# 13. Roman to Integer
def romanToInt(s):
    roman = {
        'I': 1, 'V': 5, 'X': 10,
        'L': 50, 'C': 100,
        'D': 500, 'M': 1000
    }
    
    total = 0
    for i in range(len(s)):
        # if current value < next value â†’ subtract it
        if i + 1 < len(s) and roman[s[i]] < roman[s[i + 1]]:
            total -= roman[s[i]]
        else:
            total += roman[s[i]]
    return total


print(romanToInt("III"))
```

```python
# 14. Longest Common Prefix

words = ["flower", "flow", "flight"]

# Start with the first word as your starting point
prefix = words[0]
print("Prefix: "+prefix)

# Compare the prefix with each next word
for word in words[1:]:

    print("word:" +word)   # compares "flight". startwith("flower") and then "flight".startwith("flowe")
    while not word.startswith(prefix):
        prefix = prefix[:-1]  # remove the last letter
        if prefix == "":
            break  # stop if no common prefix left

print("Longest common prefix:", prefix)
```

```python
# 15. 3Sum
nums = [-1, 0, 1, 2, -1, -4]
ListOfList = []
unique = set()

for a in range(len(nums)):
    for b in range(a + 1, len(nums)):
        for c in range(b + 1, len(nums)):
            if nums[a] + nums[b] + nums[c] == 0:
                triplet = tuple(sorted([nums[a], nums[b], nums[c]]))
                if triplet not in unique:
                    unique.add(triplet)
                    ListOfList.append(list(triplet))

print(ListOfList)
```

```python

# 16. 3Sum Closest
nums = [-1, 2, 1, -4]
target = 1

dic = {}

# Store each triplet and its sum
for a in range(len(nums)):
    for b in range(a + 1, len(nums)):
        for c in range(b + 1, len(nums)):
            total = nums[a] + nums[b] + nums[c]
            triplet = tuple(sorted([nums[a], nums[b], nums[c]]))
            dic[triplet] = total

# Find the sum closest to the target
closest_sum = list(dic.values())[0]
for val in dic.values():
    if abs(val - target) < abs(closest_sum - target):
        closest_sum = val

print("All triplet sums:", dic)
print("Closest sum:", closest_sum)


```

```python
# 20. Valid Parentheses

def isValid(s):
    open_brackets = ['(', '{', '[']
    close_brackets = [')', '}', ']']
    
    stack = []   

    for ch in s:  
        if ch in open_brackets:   
            stack.append(ch)     
        
        elif ch in close_brackets:  
            if not stack:            
                return False         

            last = stack.pop()      
            open_index = open_brackets.index(last)  
            correct_close = close_brackets[open_index]  

            if ch != correct_close:  
                return False          

    return len(stack) == 0  

s = "([])"
isValid(s)
```

```python
#E 26. Remove Duplicates from Sorted Array
nums = [0,0,1,1,1,2,2,3,3,4]
ListLength=len(nums)
CleanList= []
for i in nums:
    if i not in CleanList:
        CleanList.append(i)

CleanList= sorted(CleanList)
PlaceholdersNeeded= ListLength-len(CleanList)
print(len(CleanList))
for j in range(PlaceholdersNeeded):
    CleanList.append('_')
    
print(CleanList)
```
